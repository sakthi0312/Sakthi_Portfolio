```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Calories'], bins=20, kde=True)
plt.title('Distribution of Calorie Counts Across Menu Items')
plt.xlabel('Calories')
plt.ylabel('Item')
plt.show()
```

A right-skewed distribution is also referred to as positively skewed

it implies that the majority of the data points are concentrated on the left side of the distribution, with a tail extending to the right ,The right tail of the distribution is longer, indicating the presence of relatively few data points with higher values.

These higher values contribute to the skewness and can affect the interpretation of central tendency.

So now we have to identify the outlier which causing the skewness and handle it

```python
#identifying outlier with boxplot
sns.boxplot(x=df['Calories'])
```

we can understand from above boxplot that there are some outliers in the data above 900 calories (approx)
so we are going to handle it

```python
#making a copy and displaying those outliers
df_copy=df.copy()
df[df['Calories']>900]
```

```python
#dropping the outliers
df_copy.drop(df[df['Calories']>900].index,inplace = True)
```

# Data Visualization

```python
#preparing a count plot to display the no.of items in specific categories
sns.set(font_scale=1)
plt.figure(figsize=(10, 5))
sns.countplot(y='Category', data=df)
```

```python
#getting the distribution of calories and cholesterol over categories

f, ax = plt.subplots(1, 2, figsize=(25, 12))
MeanCalories = {}
MeanCholesterol = {}
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Calories'].mean())
    MeanCalories[category] = mean
pd.Series(MeanCalories).sort_values(ascending=True).plot.barh(width=0.8, color=sns.color_palette('RdYlGn',10), ax=ax[0])
ax[0].set_title("Mean Calories of Different Categories")

for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Cholesterol (% Daily Value)'].mean())
    MeanCholesterol[category] = mean
pd.Series(MeanCholesterol).sort_values(ascending=True).plot.barh(width=0.8, color=sns.color_palette('RdYlGn',10), ax=ax[1])
ax[1].set_title("Mean Cholesterol of Different Categories")
plt.subplots_adjust(wspace=0.5)
```

There are two menu options, each offering two variations of eggs (egg whites only and whole eggs). We compare four menus in total. The bar charts reveal that dishes containing whole eggs generally have higher total fat, saturated fat, cholesterol, and other nutritional components. This is logical since whole eggs include egg yolks, which have higher levels of fat, cholesterol, and protein. Opting for a menu with egg whites would be a suitable choice if you prefer an egg-based dish with lower fat and cholesterol while maintaining similar nutritional values. On the other hand, if you seek the original nutritional profile of eggs, selecting a menu that incorporates whole eggs in its ingredients would be more appropriate.

```python
# creating a function to generate a interactive plot by category and a table containing all food and nutrients of that catagory

def visualize_category(category):
    
    # items ascending=False
    items = df.query('Category == @category').sort_values(by="Calories", ascending=False)
    items_custom = items.iloc[:, [1, 3, 5]]

    # items ascending=True
    items_t = df.query('Category == @category').sort_values(by="Calories", ascending=True)
    items_custom_t = items_t.iloc[:, [1, 3, 5]]

    # Print Category as title
    print(f"Visualizing category: {category}")

    # interactive chart with plotly express
    p_express = px.scatter(items_custom, x="Total Fat", y='Calories', color="Item", size='Calories', template="plotly_dark",
                            hover_name="Item")
    p_express.show()
    
    # Display Tables food and nutrients ranked by calories
    
    table = items_custom.iloc[:, [0, 1]].sort_values(by="Calories", ascending=False).style.hide_index()
    display(table)
```

```python
#calling the function to create a visual on "Chicken & Fish"
visualize_category("Chicken & Fish")
```

```python
visualize_category("Beef & Pork")
```

```python
visualize_category("Breakfast")
```

```python
visualize_category("Smoothies & Shakes")
```
